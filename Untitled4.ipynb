```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
image_path = '/content/login.jpg'
image = Image.open(image_path)
image = image.convert('L')
histogram, bins = np.histogram(image, bins=256, range=(0, 256))
cdf = histogram.cumsum()
cdf_normalized = cdf * float(histogram.max()) / cdf.max()
equalized_image = np.interp(image, bins[:-1], cdf_normalized)
equalized_image = np.uint8(equalized_image)
equalized_image_pil = Image.fromarray(equalized_image)
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')
plt.subplot(1, 2, 2)
plt.imshow(equalized_image_pil, cmap='gray')
plt.title('Equalized Image')
plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import ndimage
from PIL import Image
image_path = ''
image = Image.open("/content/login.jpg")
image = image.convert('L')
image_array = np.array(image)
kernel_size = 5
sigma = 1.0
kernel = np.fromfunction(
    lambda x, y: (1/ (2 * np.pi * sigma ** 2)) *
                  np.exp(-((x - (kernel_size - 1) / 2) ** 2 +
                           (y - (kernel_size - 1) / 2) ** 2) / (2 * sigma ** 2)),
    (kernel_size, kernel_size)
)
kernel = kernel / np.sum(kernel)
smoothed_image = ndimage.convolve(image_array, kernel, mode='constant', cval=0.0)
smoothed_image_pil = Image.fromarray(np.uint8(smoothed_image))
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')
plt.subplot(1, 2, 2)
plt.imshow(smoothed_image_pil, cmap='gray')
plt.title('Smoothed Image (Gaussian Filter)')
plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
image_path = '/content/login.jpg'
image = Image.open(image_path)
image = image.convert('L')
image_array = np.array(image, dtype=float)
dx = np.gradient(image_array, axis=1)
dy = np.gradient(image_array, axis=0)
gradient_magnitude = np.sqrt(dx**2 + dy**2)
gradient_magnitude = (gradient_magnitude / gradient_magnitude.max()) * 255
gradient_image = Image.fromarray(np.uint8(gradient_magnitude))
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')
plt.subplot(1, 2, 2)
plt.imshow(gradient_image, cmap='gray')
plt.title('First Order Derivative (Gradient)')
plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
image_path = '/content/login.jpg'
image = Image.open(image_path)
image = image.convert('L')
image_array = np.array(image, dtype=float)
laplacian_kernel = np.array([[0, 1, 0], [1, -4, 1], [0, 1, 0]])
second_derivative = np.abs(np.fft.ifft2(np.fft.fft2(image_array) * np.fft.fft2(laplacian_kernel, s=image_array.shape)))
second_derivative = (second_derivative / second_derivative.max()) * 255
second_derivative_image = Image.fromarray(np.uint8(second_derivative))
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')
plt.subplot(1, 2, 2)
plt.imshow(second_derivative_image, cmap='gray')
plt.title('Second Order Derivative (Laplacian)')
plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
from PIL import Image, ImageDraw

def find_edges(image):
    grayscale_image = image.convert("L")
    edge_image = Image.new("L", image.size, 0)
    kernel_x = [[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]]
    kernel_y = [[-1, -2, -1], [0, 0, 0], [1, 2, 1]]
    width, height = image.size

    for y in range(1, height - 1):
        for x in range(1, width - 1):
            Gx = sum(sum(kernel_x[i][j] * grayscale_image.getpixel((x + j - 1, y + i - 1))
                            for j in range(3)) for i in range(3))
            Gy = sum(sum(kernel_y[i][j] * grayscale_image.getpixel((x + j - 1, y + i - 1))
                            for j in range(3)) for i in range(3))
            gradient_magnitude = int((Gx * 2 + Gy * 2) ** 0.5)
            edge_image.putpixel((x, y), gradient_magnitude)

    return edge_image

def draw_border(image, edge_color, border_thickness):
    draw = ImageDraw.Draw(image)
    width, height = image.size

    for y in range(height):
        for x in range(width):
            if image.getpixel((x, y)) > 0:
                for i in range(border_thickness):
                    draw.rectangle([x - i, y - i, x + i, y + i], outline=edge_color)

if __name__ == "__main__":
    input_image_path = "/content/login.jpg"  # Replace with the path to your input image
    output_image_path = "/path/to/output_image.jpg"
 # Replace with the desired output path
    input_image = Image.open(input_image_path)
    edge_image = find_edges(input_image)
    draw_border(edge_image, edge_color=255, border_thickness=2)
    edge_image.show()
    edge_image.save(output_image_path)

```
