```python
#6. Identifying face in an image
#6a. Identify and mark eyes
#6b. Identify and mark eyebrows
#6c. Identify and mark cheeks
#6d. Identify and mark lips
#6e. Identify and mark fire head


import urllib.request
import cv2
from google.colab.patches import cv2_imshow
# Define the URL to the Haar Cascade XML file
cascade_url = "https://raw.githubusercontent.com/opencv/opencv/master/data/haarcascades/haarcascade_frontalface_default.xml"

# Download the XML file
urllib.request.urlretrieve(cascade_url, "haarcascade_frontalface_default.xml")

# Load the face detection model
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')



 # 6a. Identify and mark eyes
import cv2
eye_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_eye.xml')
# Load the image
image = cv2.imread('/content/download.png')
# Convert the image to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
# Detect eyes in the image
eyes = eye_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))
# Draw rectangles around both eyes
for (x, y, w, h) in eyes:
    cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 2)
# display the result
cv2_imshow(image)
```

```python
#6b. Identify and mark eyebrows
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
# Load the input image
input_image_path = '/content/download.png'
image = cv2.imread(input_image_path)
# Convert the image to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
faces = face_cascade.detectMultiScale(gray, scaleFactor=1.3, minNeighbors=5)
for (x, y, w, h) in faces:
    # Extract the region of interest (ROI) around the face
    roi = image[y:y + h, x:x + w]
    # Define the region for eyebrows (slightly higher than the eyes)
    eyebrow_y = y + int(h * 0.25)
    eyebrow_region = roi[eyebrow_y:eyebrow_y + int(h * 0.1), x:x + w]
    # Draw a rectangle around the eyebrows on the original image
    cv2.rectangle(image, (x, eyebrow_y), (x + w, eyebrow_y + int(h * 0.1)), (0, 255, 0), 2)
# Display the image with marked eyebrows
cv2_imshow(image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
 #6c. Identify and mark cheeks
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
# Load the input image
input_image_path = 'img2.jpeg'
image = cv2.imread(input_image_path)
# Convert the image to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
faces = face_cascade.detectMultiScale(gray, scaleFactor=1.3, minNeighbors=5)
for (x, y, w, h) in faces:
    # Extract the region of interest (ROI) around the face
    roi = image[y:y + h, x:x + w]
    # Define the region for the left cheek (higher on the face)
    left_cheek_x = x
    left_cheek_y = y + int(h * 0.4)
    left_cheek_width = int(w * 0.4)
    left_cheek_height = int(h * 0.4)
    left_cheek = roi[left_cheek_y:left_cheek_y + left_cheek_height, left_cheek_x:left_cheek_x + left_cheek_width]
    # Draw a rectangle around the left cheek on the original image
    cv2.rectangle(image, (left_cheek_x, left_cheek_y), (left_cheek_x + left_cheek_width, left_cheek_y + left_cheek_height), (0, 255, 0), 2)
    # Define the region for the right cheek (higher on the face)
    right_cheek_x = x + int(w * 0.6)
    right_cheek_y = y + int(h * 0.4)
    right_cheek_width = int(w * 0.4)
    right_cheek_height = int(h * 0.4)
    right_cheek = roi[right_cheek_y:right_cheek_y + right_cheek_height, right_cheek_x:right_cheek_x + right_cheek_width]
    # Draw a rectangle around the right cheek on the original image
    cv2.rectangle(image, (right_cheek_x, right_cheek_y), (right_cheek_x + right_cheek_width, right_cheek_y + right_cheek_height), (0, 255, 0), 2)
# Display the image with marked left and right cheeks
cv2_imshow(image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#6d. Identify and mark lips
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
# Load the input image
input_image_path = 'img.jpeg'
image = cv2.imread(input_image_path)
# Convert the image to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
faces = face_cascade.detectMultiScale(gray, scaleFactor=1.3, minNeighbors=5)
for (x, y, w, h) in faces:
    # Extract the region of interest (ROI) around the face
    roi = image[y:y + h, x:x + w]
    # Define the region for the lips
    lips_x = x + int(w * 0.3)
    lips_y = y + int(h * 0.7)
    lips_width = int(w * 0.4)
    lips_height = int(h * 0.1)
    lips = roi[lips_y:lips_y + lips_height, lips_x:lips_x + lips_width]
    # Draw a rectangle around the lips on the original image
    cv2.rectangle(image, (lips_x, lips_y), (lips_x + lips_width, lips_y + lips_height), (0, 255, 0), 2)
# Display the image with marked lips
cv2_imshow(image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#6e. Identify and mark fire head
import dlib
import cv2
from google.colab.patches import cv2_imshow
# Load the input image
input_image_path = 'img.jpeg'
image = cv2.imread(input_image_path)
detector = dlib.get_frontal_face_detector()
predictor = dlib.shape_predictor('shape_predictor_68_face_landmarks.dat')
# Convert the image to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
# Detect faces in the grayscale image
faces = detector(gray)
for face in faces:
    landmarks = predictor(gray, face)
    # Extract forehead region using facial landmarks and make it larger
    forehead_points = [landmarks.part(17), landmarks.part(21), landmarks.part(22), landmarks.part(26)]
    forehead_x = min(point.x for point in forehead_points)
    forehead_y = max(0, min(point.y for point in forehead_points) - 30)
    forehead_width = max(point.x for point in forehead_points) - forehead_x
    forehead_height = max(10, min(point.y for point in forehead_points) - forehead_y)
    # Draw a rectangle around the forehead on the original image
    cv2.rectangle(image, (forehead_x, forehead_y), (forehead_x + forehead_width, forehead_y + forehead_height), (0, 255, 0), 2)
# Display the image with the forehead box marked
cv2_imshow(image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
